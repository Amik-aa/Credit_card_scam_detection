Importing required libraries and packages

```python
import numpy as np
import pandas as pd
```

Context:
It is important that credit card companies are able to recognize fraudulent credit card transactions so that customers are not charged for items that they did not purchase.

Content:
This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.
The features represent various aspects of the transaction, and the class label indicates whether the transaction was fraudulent (class 1) or not (class 0).

```python
# Dataset link:  https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud
```

```python
data = pd.read_csv("creditcard.csv")
```

Data is not balanced in the column named "Class"  for which we want to classify transactions, to handle this problem we'll create a sample as a dataframe of those values which are balanced and going to be used for model building and training purpose.

```python
N_fraud = data[data.Class==0]
```

Dataframe for transactions which are not fraudulent

```python
fraud = data[data['Class']==1]
```

Dataframe for transactions which are fraudulent

```python
N_fraud
```

```python
fraud
```

```python
N_fraud_sample = N_fraud.sample(n=492)
```

Here we created a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

```python
N_fraud_sample['Class'].value_counts()
```

```python
fraud['Class'].value_counts()
```

```python
new_df = pd.concat([N_fraud_sample,fraud],axis=0)
```

```python
new_df
```

```python
new_df['Class'].value_counts()
```

splitting data for classification

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
X
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=28)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
```

training the Logistic Regression Model with Training Data

```python
 model.fit(X_train, y_train)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
print('Accuracy : ', test_data_accuracy)
```
